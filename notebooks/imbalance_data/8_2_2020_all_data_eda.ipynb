```python
def find_delta_from_350_to_355(input_df, stock_symbol_dict):
    first_year = None
    first_day = None
    first_month = None
    stock_number  = None
    stock_symbol_to_difference_dict = {
        'symbol':[], 'diff_notional_imbalance':[], 'diff_lastprice':[] , 'diff_nbb':[] , 'diff_nbo':[] , 'diff_imbalancesize':[], 'diff_notional_imbalance':[], 'diff_matchedsize': [], 'diff_currentreferenceprice':[], 'diff_nearindicativeprice': [] , 'diff_farindicativeprice':[], 'diff_matched_to_unmatched':[], 'diff_currentvolume': [] , 'diff_imbside': [], 'symbol': [], 'year': [] , 'month': [], 'day': []
        }
    original_355pm_array_of_df = []

    for stock_name,stock_num in stock_symbol_dict.items():
        individual_stock_df = input_df[input_df.symbol == stock_num]
        stock_number = stock_num
        # match year  month and day
        for rows in individual_stock_df.iterrows():
            row_data = rows[1]
            if first_year == None:
                first_year = row_data['year']
                first_day =  row_data['day']
                first_month =  row_data['month']

                # variables that should change
                first_notional_imbalance = row_data['notional_imbalance']
                first_last_price= row_data['lastprice']
                first_nbb= row_data['nbb']
                first_nbo= row_data['nbo']
                first_imbalancesize = row_data['imbalancesize']
                first_notional_imbalance= row_data['notional_imbalance']
                first_matchedsize = row_data['matchedsize']
                first_currentreferenceprice = row_data['currentreferenceprice']
                first_nearindicativeprice = row_data['nearindicativeprice']
                first_farindicativeprice= row_data['farindicativeprice']
                first_matched_to_unmatched = row_data['matched_to_unmatched']
                first_currentvolume = row_data['currentvolume']
                first_imbside = row_data['imbside']

            else: # the next row should match verify with dates. Build new DF with this data (355pm snapshot AND 350 - 355pm difference)
                if (row_data['year'] == first_year) and (row_data['day'] == first_day) and ( row_data['month'] == first_month):
                    stock_symbol_to_difference_dict['symbol'].append(stock_num)

                    # append the original 355pm data
                    original_355pm_array_of_df.append(pd.DataFrame(row_data))
                
                # variables that change (1) last price (2) nbb (3)  nbo (4)imbalancesize 
                # (5)notional_imbalance (6)matchedsize (7)   currentreferenceprice (8) nearindicativeprice (9) farindicativeprice (10) matched_to_unmatched (11) currentvolume (12) imbside
                    # diff is new minus old
                    stock_symbol_to_difference_dict['diff_notional_imbalance'].append(row_data['notional_imbalance'] - first_notional_imbalance )
                    stock_symbol_to_difference_dict['diff_lastprice'].append(row_data['lastprice'] - first_last_price)
                    stock_symbol_to_difference_dict['diff_nbb'].append(row_data['nbb'] - first_nbb)
                    stock_symbol_to_difference_dict['diff_nbo'].append(row_data['nbo'] - first_nbo)
                    stock_symbol_to_difference_dict['diff_imbalancesize'].append(row_data['imbalancesize'] - first_imbalancesize )
                    stock_symbol_to_difference_dict['diff_matchedsize'].append(row_data['matchedsize'] - first_matchedsize )
                    stock_symbol_to_difference_dict['diff_currentreferenceprice'].append(row_data['currentreferenceprice'] - first_currentreferenceprice )
                    stock_symbol_to_difference_dict['diff_nearindicativeprice'].append(row_data['nearindicativeprice'] - first_nearindicativeprice )
                    stock_symbol_to_difference_dict['diff_farindicativeprice'].append(row_data['farindicativeprice'] - first_farindicativeprice)
                    stock_symbol_to_difference_dict['diff_matched_to_unmatched'].append(row_data['matched_to_unmatched'] - first_matched_to_unmatched )
                    stock_symbol_to_difference_dict['diff_currentvolume'].append(row_data['currentvolume'] - first_currentvolume )
                    stock_symbol_to_difference_dict['diff_imbside'].append(row_data['imbside'] - first_imbside )
                    stock_symbol_to_difference_dict['year'].append(row_data['year'])
                    stock_symbol_to_difference_dict['month'].append(row_data['month'])
                    stock_symbol_to_difference_dict['day'].append(row_data['day'])

                    # reset the stock data we are looking for 

                    first_year = None
                    first_day= None
                    first_month = None
                    stock_number = None
    original_355_df = pd.concat(original_355pm_array_of_df,axis=1).T
    stock_symbol_diff_df = pd.DataFrame(stock_symbol_to_difference_dict)
    return pd.merge( original_355_df, stock_symbol_diff_df,  how='left', left_on=['year','month', 'day', 'symbol'], right_on = ['year','month', 'day', 'symbol'])
```

# Normalize the cols
- mean near 0, var near 1

```python
def normalize_col(input_col):
    max_ = input_col.max()
    min_ = input_col.min()
    return (input_col - min_) / (max_ - min_)

def standardize(input_col):
    return (input_col - np.mean(input_col)) / np.std(input_col)

def normalize_and_standardize(input_col):
    n_col = normalize_col(input_col)
    return standardize(n_col)
```

```python
def normalize_df(input_df,cols_to_not_normalize=['close','pct_move_snapshot_to_close']):
    # ensure we dot NOT normalize the y col
    new_df = input_df.copy()
    for col in new_df.columns:
        print(col)
        if col == Y_COL:
            continue
        elif col == 'close': # need to drop close from our model!
            continue
        elif col == 'lastprice':
            new_df[f"{col}_unnormalized"] = new_df[f"{col}"]
        normalized_c = normalize_and_standardize(new_df[f"{col}"])
        new_df[f"{col}"] = normalized_c
        print(new_df[f"{col}"] .mean(), new_df[f"{col}"] .var())
    return new_df
```
